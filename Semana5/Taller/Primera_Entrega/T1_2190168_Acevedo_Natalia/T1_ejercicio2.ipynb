```python
import numpy as np
import sympy as sp
from math import atan, cos, sin
```

```python
# MEDIDAS INICIALES

# Longitudes de los elementos en m
Longitud_AB = 2
Longitud_DE, Longitud_BC = Longitud_AB, Longitud_AB
Longitud_AD = 1
Longitud_BE = Longitud_AD

# Ángulos entre elementos 
Angulo_BCE = atan(Longitud_BE/Longitud_BC)
Angulo_BAE = Angulo_BCE

# Ángulos de la fuerza P
Teta_inicial = 0
Teta_final1 = 90
Teta_final2 = 180
Aumento_teta = 10

# Fuerza P en kN
P_base = 20
Aumento_P = 0.25
P_max1 = 20 + (Teta_final1/Aumento_teta)*Aumento_P
P_max2 = 20 + (Teta_final2/Aumento_teta)*Aumento_P
```

```python
# Rango 1 de ángulos de la fuerza P
rango_teta1 = np.arange(Teta_inicial, Teta_final1, Aumento_teta)
rango_P1 = np.arange(P_base, P_max1, Aumento_P)
rango_P1
```

```python
# EQUILIBRIO EXTERNO 1
Ax, Ay, Dx = sp.symbols("Ax Ay Dx")

def reacciones(carga1, angulo1):
    # Fy = Ay + carga1*cos(angulo1*np.pi / 180)
    Ay = -carga1*cos(angulo1*np.pi / 180)
    
    # Ma = Longitud_AD*Dx + 2*Longitud_AB*carga1*cos(angulo1*np.pi / 180)
    Dx = (-2*Longitud_AB*carga1*cos(angulo1*np.pi / 180))/Longitud_AD
    
    # Fx = Dx + Ax + carga1*sin(angulo1*np.pi / 180)
    Ax = -Dx -carga1*sin(angulo1*np.pi / 180)
    return Ax, Ay, Dx
```

```python
# Nodo C
def Nodo_C(carga1, angulo1):
    # Sumatoria de fuerzas en y
    # Fy = carga1*cos(angulo1*np.pi / 180) - EC*sin(Angulo_BAE) = 0
    EC = (carga1*cos(angulo1*np.pi / 180))/sin(Angulo_BAE)
    # Sumatoria de fuerzas en x
    # Fx = carga1*sin(angulo1*np.pi / 180) - EC*cos(Angulo_BAE) - BC= 0
    BC = carga1*sin(angulo1*np.pi / 180) - EC*cos(Angulo_BAE)
    return EC, BC

# Nodo D
def Nodo_D(Dx):
    # Sumatoria de fuerzas en x
    # Fx = Dx + DE = 0
    DE = -Dx
    # Sumatoria de fuerzas en y
    AD = 0
    return DE, AD

# Nodo E
def Nodo_E(EC):
    # Sumatoria de fuerzas en y
    # Fy = AE*sin(Angulo_BAE) + EC*sin(Angulo_BAE) = 0
    AE = -EC
    # Sumatoria de fuerzas en X
    # Fx = -AE*cos(Angulo_BAE) + EC*cos(Angulo_BAE) - ED = 0
    ED = -AE*cos(Angulo_BAE) + EC*cos(Angulo_BAE)
    return AE, ED

# Nodo A
def Nodo_A(Ay, Ax):
    # Sumatoria de fuerzas en y
    # Fy = EA*sin(Angulo_BAE) - Ay = 0
    EA = Ay/sin(Angulo_BAE)
    # Sumatoria de fuerzas en X
    # Fx = Ax + EA*cos(Angulo_BAE) + AB = 0
    AB = - EA*cos(Angulo_BAE) - Ax
    return EA, AB

#Nodo B
def Nodo_B(AB):
    # Sumatoria de fuerzas en x
    # Fx = -AB + CB = 0
    CB = AB
    # Sumatoria de fuerzas en y
    BE = 0
    return CB, BE
```

```python
for angulo1, carga1 in zip(rango_teta1, rango_P1):
    Ax, Ay, Dx = reacciones(carga1, angulo1)
    EC, BC = Nodo_C(carga1, angulo1)
    DE, AD = Nodo_D(Dx)
    AE, ED = Nodo_E(EC)
    EA, AB = Nodo_A(Ay, Ax)
    CB, BE = Nodo_B(AB)
    print(f"El valor de las reacciones para un ángulo de {angulo1} grados medido desde el eje Y positivo: ")
    print(Ax, Ay, Dx)
    print("El valor de las fuerzas en los elementos:")
    print(AB, AE, AD, BC, BE, EC, DE)
    print(f"{EA} == {AE}, {ED} == {DE}, {CB} == {BC}")
    print()
```

```python
# Rango 2 de ángulos de la fuerza P
#Esta relación se pone para que los ángulos se sigan trabajando desde la vertical
rango_teta2 = 180 - np.arange(Teta_final1, Teta_final2 + 10, Aumento_teta)
rango_P2 = np.arange(P_max1, P_max2 + Aumento_P, Aumento_P)
rango_P2
```

```python
# EQUILIBRIO EXTERNO 2
Ax, Ay, Dx = sp.symbols("Ax Ay Dx")

def reacciones(carga2, angulo2):
    # Fy = Ay - carga2*cos(angulo2*np.pi / 180)
    Ay = carga2*cos(angulo2*np.pi / 180)
    
    # Ma = Longitud_AD*Dx - 2*Longitud_AB*carga2*cos(angulo2*np.pi / 180)
    Dx = (2*Longitud_AB*carga2*cos(angulo2*np.pi / 180))/Longitud_AD
    
    # Fx = Dx + Ax + carga2*sin(angulo2*np.pi / 180)
    Ax = - Dx - carga2*sin(angulo2*np.pi / 180)
    return Ax, Ay, Dx
```

```python
# Nodo C
def Nodo_C(carga2, angulo2):
    # Sumatoria de fuerzas en y
    # Fy = - carga2*cos(angulo2*np.pi / 180) - EC*sin(Angulo_BAE) = 0
    EC = (-carga2*cos(angulo2*np.pi / 180))/sin(Angulo_BAE)
    # Sumatoria de fuerzas en x
    # Fx = carga2*sin(angulo2*np.pi / 180) - EC*cos(Angulo_BAE) + BC= 0
    BC = carga2*sin(angulo2*np.pi / 180) - EC*cos(Angulo_BAE)
    return EC, BC

# Nodo D
def Nodo_D(Dx):
    # Sumatoria de fuerzas en x
    # Fx = Dx + DE = 0
    DE = - Dx
    # Sumatoria de fuerzas en y
    AD = 0
    return DE, AD

# Nodo E
def Nodo_E(EC):
    # Sumatoria de fuerzas en y
    # Fy = AE*sin(Angulo_BAE) + EC*sin(Angulo_BAE) = 0
    AE = -EC
    # Sumatoria de fuerzas en X
    # Fx = AE*cos(Angulo_BAE) - EC*cos(Angulo_BAE) + ED = 0
    ED = - AE*cos(Angulo_BAE) + EC*cos(Angulo_BAE)
    return AE, ED

# Nodo A
def Nodo_A(Ay, Ax):
    # Sumatoria de fuerzas en y
    # Fy = - EA*sin(Angulo_BAE) + Ay = 0
    EA = Ay/sin(Angulo_BAE)
    # Sumatoria de fuerzas en X
    # Fx = Ax + EA*cos(Angulo_BAE) + AB = 0
    AB = - EA*cos(Angulo_BAE) + - Ax
    return EA, AB

#Nodo B
def Nodo_B(AB):
    # Sumatoria de fuerzas en x
    # Fx = AB - CB = 0
    CB = AB
    # Sumatoria de fuerzas en y
    BE = 0
    return CB, BE
```

```python
for angulo2, carga2 in zip(rango_teta2, rango_P2):
    Ax, Ay, Dx = reacciones(carga2, angulo2)
    EC, BC = Nodo_C(carga2, angulo2)
    DE, AD = Nodo_D(Dx)
    AE, ED = Nodo_E(EC)
    EA, AB = Nodo_A(Ay, Ax)
    CB, BE = Nodo_B(AB)
    print(f"El valor de las reacciones para un ángulo de {angulo2} grados medido desde el eje Y negativo: ")
    print(Ax, Ay, Dx)
    print("El valor de las fuerzas en los elementos:")
    print(AB, AE, AD, BC, BE, EC, DE)
    print(f"{EA} == {AE}, {ED} == {DE}, {CB} == {BC}")
    print()
```
